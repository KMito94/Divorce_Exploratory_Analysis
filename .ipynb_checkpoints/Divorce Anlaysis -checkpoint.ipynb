 ## Exploratory Analysis of Divorce Data 

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

import pingouin as pg

%matplotlib inline
```

```python
data = pd.read_csv('divorce.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# we need to convert the data columns to data_time objcets 
```

```python
data['divorce_date'] = pd.to_datetime(data['divorce_date'])
```

```python
data['dob_man'] = pd.to_datetime(data['dob_man'])
```

```python
data['dob_woman'] = pd.to_datetime(data['dob_woman'])
```

```python
data['marriage_date'] = pd.to_datetime(data['marriage_date'])
```

```python
data.dtypes
```

```python
# the percentage of missing values 
```

```python
percent_of_missing_values = (data.isnull().sum()/len(data))*100
```

```python
percent_of_missing_values
```

```python
# The null values are more than 30 percent so we need to fill them 
```

```python
# we fill the missing values in the num_kinds column with the mean 
```

```python
data['num_kids'].fillna(data['num_kids'].mean(),inplace=True)
```

```python
data.info()
```

```python
# We have all the values that we need 
```

```python
data.head()
```

```python
plt.style.use('ggplot')
```

```python
#The level of education among men

plt.figure(figsize=(8,6))

sns.countplot(x='education_man',data=data);

plt.xticks(rotation=60);
```

```python
# The level of education among women 
```

```python
plt.figure(figsize=(8,6))

sns.countplot(x='education_woman',data=data);

plt.xticks(rotation=60);
```

```python
# compare marriage duration and the number of kinds 
```

```python
plt.figure(figsize=(8,6))

sns.lineplot(x='marriage_duration',y='num_kids',data=data,ci=None);
```

```python
from datetime import datetime as dt

data['divorce_year'] = data['divorce_date'].dt.year
data['marriage_year'] = data['marriage_date'].dt.year
```

```python
data.head()
```

```python
# we can determine the correlation between the duration of marriage and divorce year 
```

```python
plt.figure(figsize=(8,6))

sns.lineplot(x='marriage_year',y='marriage_duration',data=data);
```

```python
plt.figure(figsize=(8,6))

sns.barplot(x='education_woman',y='marriage_duration',data=data);
```

```python
plt.figure(figsize=(8,6))

sns.barplot(x='education_man',y='marriage_duration',data=data);
```

```python
# The level of education does not really affect the duration of marriage 
```

```python
data['marriage_year'] = data['marriage_date'].dt.year
```

```python
data['man_age_marriage'] = data['marriage_year'] - data['dob_man'].dt.year
data['woman_age_marriage'] = data['marriage_year'] - data['dob_woman'].dt.year
```

```python
data.head(20)
```

```python
data['age_difference'] = data['man_age_marriage'] - data['woman_age_marriage']
```

```python
plt.figure(figsize=(8,6))

sns.lineplot(x='marriage_duration',y='age_difference',data=data);
```

```python
plt.figure(figsize=(8,6))


sns.kdeplot(x='marriage_duration',data=data,hue='education_man',cut=0);
```

```python
sns.relplot(x='woman_age_marriage',y='num_kids',kind='line',data=data);
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x='man_age_marriage',y='woman_age_marriage',data=data,hue='education_man');
```

```python
mean_man_income_profession = data.groupby('education_man')['income_man'].mean()
```

```python
mean_man_income_profession
```

```python
plt.figure(figsize=(8,6))

mean_man_income_profession.plot(kind='bar',color='green')

plt.xlabel("Level of Man Education")

plt.ylabel('Icome')

plt.title('Comparion of Man Income and Levle of Education');
```
